**using groupby function to organize random data in to groups**

```python
import pandas as pd
var1 = {'a':[1,2,3,4,5,6,7,8,9,0],
       'b':[6,7,8,9,0,1,2,3,4,5],
        'c':['a','b','c','d','e','f','g','h','i','j']}
dt_var1 = pd.DataFrame(var1)
# print(dt_var1)
var_new = dt_var1.groupby('a')
print(var_new)
var_new.get_group(0)
var_new.max()
var_new.min()
var_new.mean() # it works only for numerical value
```

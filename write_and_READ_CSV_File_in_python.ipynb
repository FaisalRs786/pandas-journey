**WRITE CSV FILE USING TO_CVS FUNCTION**

```python
# HERE I CREATE CSV FILE FROM DATAFRAME
import pandas as pd
dic_var = {'languages':['python','java','c','c++','kotlin','c#','javascript','php'],
           'rankings':[1,2,8,6,5,7,3,4],
           'popularity':[1,4,5,6,8,2,3,7],
           'salary':[100,200,300,400,500,600,700,800]}
dt_frame = pd.DataFrame(dic_var)
dt_frame.name='faisal'
# our data frame is ready
print(dt_frame)

# now convert this datframe in to csv file
# ........func....name of csv file............
dt_frame.to_csv("testcsv",index=False)

```

**READ CSV FILE**



```python
# here i read csv file using pd.read_csv function
import pandas as pd
csv_file_var = pd.read_csv('/content/testcsv')
csv_file_var
# print(csv_file_var)
```

```python
# we can get particular row of data set as using [ nrows ]parameter
csv_file_var1 = pd.read_csv('/content/testcsv', nrows =2) #it shows on two upper rows
print(csv_file_var1)

print("""

........................
                           """)
# we can get particular column using [usecols] parameter as
# .........................................................column name, column name....
csv_file_var2 = pd.read_csv('/content/testcsv', usecols = ['languages','salary'])
print(csv_file_var2)

print("""

........................
                           """)

# we can also use indexes of column to call instead of name of column
csv_file_var3 = pd.read_csv('/content/testcsv', usecols = [0,2])
print(csv_file_var3)

print("""

........................
                           """)

# we can skip rows of data set as using [skiprows] parameter as
csv_file_var4 = pd.read_csv('/content/testcsv', skiprows = [2,3])
print(csv_file_var4)

print("""

........................
                           """)

# we can make any column as index column using [index_col] parameter
csv_file_var5 = pd.read_csv('/content/testcsv', index_col = 'languages')
print(csv_file_var5)

print("""

........................
                           """)

# we can make any row as header of csv file using [header] parameter
csv_file_var6 = pd.read_csv('/content/testcsv', header = 2)
print(csv_file_var6)

print("""

........................
                           """)

# we can change heading of csv file using [name] parameter
csv_file_var7 = pd.read_csv('/content/testcsv', names = ['a','b','c','d'])
print(csv_file_var7)
```

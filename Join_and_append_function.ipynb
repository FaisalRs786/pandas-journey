**join function used to join two data frames as**

```python
import pandas as pd
import pandas as pd
var1 = {'a':[1,2,3,4,5],
        'b':[6,7,8,9,0]}
var2 = {'c':[11,12,13,14,15],
        'd':[16,17,18,19,20]}
dt_var1 = pd.DataFrame(var1)
dt_var2 = pd.DataFrame(var2)
print(dt_var1)
print(dt_var2)
print(dt_var1.join(dt_var2))

# use index parameter to change index names as
dt_var1.join(dt_var2,how='right')

```

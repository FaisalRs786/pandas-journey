**REPLACE FUNCTION IS USED TO REPLACE VALUE BY OTHER VALUE**

```python
import pandas as pd
csv_var = pd.read_csv('/content/csv_test.csv')
csv_var
```

```python
# use replace function to replace values as
# .....................what value to replace......this comes at place of python
print(csv_var.replace(to_replace='python' ,value = 'kingpython'))

# we can replace multiple values as
print(csv_var.replace([1,2,3,4,5], 100)) # it will replace 1,2,3,4,5 by 100

# we use method parameter like this
print(csv_var.replace([100], method = 'ffill'))

# we can use regex parameter as to replace values as
print(csv_var.replace("[a-z]","!",regex=True))

# we can replace particular column data as
print(csv_var.replace({'ranking':1}, 1000 ))

# we can also use limit parameter to set limit and inplace paramet to make changes in orignal data set
```

**INTERPOLATE FUNCTION**

```python
# interpolate function is used to fill null data automatically according to differnert methods by default method id linear
print(csv_var.interpolate())
# here by default method is linear we can also set method = time , index , value and many more
# we can also use limit parameter to set limit and axis parameter to set by row and column
# we also set limit_direction to set start filling of null value (backward,forward,Both)
```
